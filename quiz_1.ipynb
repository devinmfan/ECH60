**PROBLEM 3-a:** The population $P(t)$ as a function of time for many microbes can be modeled by the following equation: 

$$P(t)=\frac{M}{1+\left(\frac{M-P_0}{P_0}\right) e^{-rt}},$$ 

where $M$ is the maximum possible population (called the carrying capacity), $P_0$ is the initial population at time $t=0$, and $r$ is a growth rate constant.

Write a function that takes as inputs $M$ and $P_0$ and the time $t$ in seconds, and provides $P(t)$ as output. Use $r=0.008 ~s^{-1}$. 

<font color='red'> [3.5 points] </font>

```python
import numpy as np
r = 0.008
def p (m,p0,t):
    pop = m / ((1+((m-p0)/p0))*np.exp(-r*t))
    return pop
p(70,25,10)
```

Your loop should stop when the bacterial population reaches half the carrying capacity (which is $M/2=35$)
