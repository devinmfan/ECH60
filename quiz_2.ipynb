**PROBLEM 2:** Generate two arrays $x$ and $y$ using arange, both going from -2 to +2 in steps of 0.1. Then create 2D arrays using meshgrid called xgrid and ygrid. Create another 2D array called fvalues that contains values of the function 

$$f(x,y)=\exp\left[-\frac{(x^2+10y^2)}{20}\right]\sin(x)$$ 

at each point.


Then, create ***three horizontal subplots***: 

* LEFT: 30 contour lines of $f(x,y)$ in the entire range of $x$ and $y$.
* CENTER: Slice xgrid, ygrid, an fvalues appropriately to obtain the range $-1\leq x \leq 1$ and $-1\leq y \leq 1$. Display a pseudocolor version of $f(x,y)$ in this region using pcolormesh.
* RIGHT: A simple plot of $f(x,y=0)$ vs $x$. Use the appropriate slice of fvalues to generate this plot.

<font color='red'> [4.5 points]</font>


```python
x=np.arange(-2,2,.1)
y=np.arange(-2,2,.1)
xgrid, ygrid = np.meshgrid(x,y)
fvalues=np.exp((-xgrid**2+10*ygrid**2)/(20))*np.sin(xgrid)
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.contour(fvalues,30)
plt.subplot(1,3,2)
plt.pcolormesh(fvalues[:-1,:1])
plt.subplot(1,3,3)
plt.plot(fvalues[:,2],y)
plt.show()
```
